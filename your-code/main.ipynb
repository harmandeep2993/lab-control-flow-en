## Words

```python
words = ['play', 'filling', 'bar', 'theatre', 'easygoing', 'date', 'lead', 'that', 'story',  'island']
```

**Print every word in upper case**

```python
# Print words in uppercase
for word in words:   # Use "for" loop to access the elements in words list
   print(word.upper())  # Use ".upper() string method to capitalize the word"
```

**Create a new list containing only words with 5 or more letters**

```python
# List to store words with at least 5 letters
new_list = []

for word in words:
     if len(word) > 5:
        new_list.append(word)
print(new_list)
```

**Print the first word starting with "t"**

```python
# List to store words start with "t" character
words_with_t = []

for word in words:
     if word.startswith("t"):
        words_with_t.append(word)
print(words_with_t)
```

## Numbers

**Create a list containing the square of every number from 1 to 10**

```python
# Store numbers from 1 to 10 and their squares
numbers = []
squared_numbers = []

for num in range(1, 11):
    numbers.append(num)
    squared_numbers.append(num ** 2)

print("Numbers:", numbers)
print("Squared numbers:", squared_numbers)
```

**Print a list containing the square of every odd number from 1 to 10**

```python
# List to store squares of odd numbers from 1 to 10
odd_numbers = []
odd_squares = []

for num in range(1, 11):
    if num % 2 != 0:   # Check if the number is odd
        odd_numbers.append(num)
        odd_squares.append(num ** 2)

print("Odd numbers:", odd_numbers)
print("Squared odd numbers:", odd_squares)
```

**Create a list with the squares of all multiples of 8 below 1000**

```python
# Lists to store multiples of 8 and their squares below 1000
multiples_of_8 = []
squares_of_multiples_8 = []

# Iterate through numbers from 0 to 999
for num in range(1000):
    if num % 8 == 0:
        multiples_of_8.append(num)
        squares_of_multiples_8.append(num ** 2)

print("Total numbers divisible by 8 below 1000:", len(squares_of_multiples_8))
print("Multiples of 8:", multiples_of_8)
print("Squares of multiples of 8:", squares_of_multiples_8)
```

## People

```python
people = [
    {
        "name": "Juan",
        "age": 34,
        "n_kids": 2
    },
    {
        "name": "Pepe",
        "age": 27,
        "n_kids": 0
    },
    {
        "name": "Sonia",
        "age": 41,
        "n_kids": 1
    },
    {
        "name": "Lucía",
        "age": 22,
        "n_kids": 2
    },
    {
        "name": "Leo",
        "age": 55,
        "n_kids": 5
    }
]
```

**How many people are there?**

```python
# len function to find number of people in the list
len(people)
```

**How many people have kids**?

```python
# Count how many people have at least one child
count = 0

for person in people:
    # Check if the person has one or more kids
    if person["n_kids"] > 0:
        count += 1

print(count)
```

**In a year's time, names ending with "a" will have an extra kid. Create a list of dictionaries with people's info in a year's time**

```python
# Pseducode

# 0. Define a list of dictionaries
# 1. Access the dictionaries in the list
# 2. Check the name -> search for the key (name)
# 3. Check key(name) ends with "a" go to step 4 otherwise go to step 5
```

```python
# List to store people's information in a year's time
next_year_list = []

for person in people:
    # Create a new dictionary for the updated person
    updated_person = {"name": person["name"],"age": person["age"]}

    # Increase number of kids if the name ends with 'a'
    if person["name"].endswith("a"):
        updated_person["n_kids"] = person["n_kids"] + 1
    else:
        updated_person["n_kids"] = person["n_kids"]

    next_year_list.append(updated_person)
for person, update_person in zip(people, next_year_list):
    print("\n",person.values())
    print(update_person.values())
```
