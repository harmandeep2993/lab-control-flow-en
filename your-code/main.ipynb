## Words

```python
words = ['play', 'filling', 'bar', 'theatre', 'easygoing', 'date', 'lead', 'that', 'story',  'island']
```

**Print every word in upper case**

```python
# Print the words in uppercase
for word in words:
   print(word.upper())
```

**Create a new list containing only words with 5 or more letters**

```python
# Create a new list
new_list = []

for word in words:
     if len(word) > 5:
        new_list.append(word)

print(new_list)
```

**Print the first word starting with "t"**

```python
# # Create a new list
new_list_1 = []

for word in words:
     if word.startswith("t"):
        new_list_1.append(word)

print(new_list_1)
```

## Numbers

**Create a list containing the square of every number from 1 to 10**

```python
# Create a list
square_list = []

# Iterate through range 1 to 10
for i in range(1, 11):
    square_list.append(i**2)

print(square_list)
```

**Print a list containing the square of every odd number from 1 to 10**

```python
# Create a empty list to store the square root of odd numbers
odd_numbers_list = []

# Iterate through numbers from 1 to 10
for i in range(1, 11):
    if i % 2 == 1:  # We can also use here i % 2 != 0 condition.
        odd_numbers_list.append(i**2)

# Print the list
print(odd_numbers_list)
```

**Create a list with the squares of all multiples of 8 below 1000**

```python
# Create a list
numbers_divisble_by_8 = []

# Iterate through numbers from 1 to 10
for i in range(1000):
    if i % 8 == 0: 
        numbers_divisble_by_8.append(i)

# Print the list
print("Totals numbers divisible by 8 below 1000: ", len(numbers_divisble_by_8))

print(numbers_divisble_by_8)
```

**In a year's time, names ending with "a" will have an extra kid. Create a list of dictionaries with people's info in a year's time**

```python
people
```

```python
# Pseducode

# 0. Define a list of dictionaries
# 1. Access the dictionaries in the list
# 2. Check the name -> search for the key (name)
# 3. Check key(name) ends with "a" go to step 4 otherwise go to step 5
```

```python
# Create a empty list
next_year_list = []

for item in people: #extract the dictionaries out of the list, one after the other
    temp_dict = {}
    temp_dict["name"] = item.get("name")
    temp_dict["age"] = item.get("age")
    if item.get("name").endswith("a"):
        temp_dict["n_kids"] = item.get("n_kids") + 1
    else:
        temp_dict["n_kids"] = item.get("n_kids")
    
    next_year_list.append(temp_dict)
print(next_year_list)
```
